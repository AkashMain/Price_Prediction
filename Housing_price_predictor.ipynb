Train-Test Splitting

```python
#For explaing sklearn function
import numpy as np
def split_train_test(data,test_ratio):
    np.random.seed(42)
    shuffled=np.random.permutation(len(data))
    test_set_size=int(len(data)*test_ratio)
    test_indices=shuffled[:test_set_size]
    train_indices=shuffled[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set,test_set=split_train_test(housing,0.2)
```

```python
print("Rows in train set:",len(train_set))
print("Rows in test set:",len(test_set))
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(housing,test_size=0.2,random_state=42)
print("Rows in train set:",len(train_set))
print("Rows in test set:",len(test_set))
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(housing,housing['chas']):
    strat_train_set=housing.loc[train_index]
    strat_test_set=housing.loc[test_index]
```

```python
housing=strat_train_set.drop('medv',axis=1)
housing_labels=strat_train_set['medv'].copy()
```

```python
housing.describe()
```

Feature Scaling
1. MinMaxscalar
2. StandardScalar

Creating pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
my_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('std_scalar',StandardScaler()), 
    
])
```

```python
housing_num_tr=my_pipeline.fit_transform(housing)
print(housing_num_tr)  #numpy array
```

## Selecting Desired Model 


```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model = LinearRegression()
#model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing_num_tr,housing_labels)
```

```python
some_data=housing.iloc[:6]
some_labels=housing_labels.iloc[:6]
prepared_data=my_pipeline.transform(some_data)
model.predict(prepared_data)
```

```python
list(some_labels)
```

Evaluating the Model

```python
from sklearn.metrics import mean_squared_error
housing_predictions=model.predict(housing_num_tr)
mse=mean_squared_error(housing_labels,housing_predictions)
rmse=np.sqrt(mse)
```

```python
rmse
```

Cross Validation techinque(better way to get less ampunt of rmse)

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,housing_num_tr,housing_labels,scoring='neg_mean_squared_error',cv=10)
rmse_scores=np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print('Scores: ',scores)
    print('Mean: ',scores.mean())
    print('Standard Deviation: ',scores.std())
```

```python
print_scores(rmse_scores)
```

## Testing the Model on Test Data

```python
x_test=strat_test_set.drop('medv',axis=1)
y_test=strat_test_set['medv'].copy()
x_test_prepared=my_pipeline.transform(x_test)
final_predictions=model.predict(x_test_prepared)
final_mse=mean_squared_error(y_test,final_predictions)
final_rmse=np.sqrt(final_mse)
```
